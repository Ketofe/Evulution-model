```python
import numpy as np
from Vicsek_payof import *
import matplotlib.pyplot as plt
```

```python
b=1.25
r=0.5
v_0=0.05
N=1000
iterations=1200
L=10

pos,theta,s=Vicsek_PD(v_0,r, N,L,iterations,b)
```

```python
#Function for counting the amount of colaberators

def C(s):
    su=0
    for k in s:
        if np.array_equal(k,np.array([1,0])):
            su=1+su
    return su/len(s)

C_list=[C(k) for k in s]
#Function for caculating the amount of order

def order(angle_list):
    return abs(sum(np.exp(angle_list*1j)) )/len(angle_list)

o=[order(k) for k in theta]

plt.plot(o,label='order')

plt.scatter([k for k in range(len(C_list)) ],C_list,label='fraction of colaborators')
#plt.xscale('log')
plt.xlabel('itetarions')
plt.legend()
plt.show()
```

```python
#Gett u
s_t=[]
for i in range(iterations):
    s_m=[]
    for index in range(N):
        if s[i][index][0]==1:
            s_m.append(1)
        else:
            s_m.append(0)
    s_t.append(s_m)
        
```

```python

fig, ax= plt.subplots(figsize=(6,6))
i=0

qv = ax.quiver(pos[i][:,0], pos[i][:,1], np.cos(theta[i][0]), np.sin(theta[i]),s_t[i],clim=[0,1])
plt.show()
```
